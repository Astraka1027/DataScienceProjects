```python
import pandas as pd
```

```python
data = pd.read_csv('Realestate.csv')
```

```python
data.head()
```

```python
# Normalize column houseAge using standard deviation and min/max to put on scale from 0-1
```

```python
data.houseAge.std()
```

```python
houseAge = data.houseAge
```

```python
norm = houseAge - houseAge.mean()/houseAge.std()
norm
```

```python
minmax = (houseAge - houseAge.min())/(houseAge.max() - houseAge.min())
```

```python
minmax
```

```python
data
```

```python
# Create new column for standardized houseAge and add to datafram
```

```python
data.houseAge = minmax
```

```python
data['houseAgeStandardized'] = norm
```
